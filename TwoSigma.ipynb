# PCA

```python
from sklearn import model_selection, decomposition, neural_network
```

```python
d_pca = decomposition.PCA(n_components = 4)
```

```python
clf_nn = neural_network.MLPRegressor(hidden_layer_sizes=(3,5,3))
```

```python
X = df2.iloc[:,2:-1]
Y = df2.y
```

```python
X.describe()
```

```python
(X.iloc[:10,:5] - X.iloc[:10,:5].mean())
```

```python
X[['derived_1']]
```

```python
d_pca.fit(X)
```

```python
Xt = d_pca.transform(X)
```

```python
k = 5 #number of folds
kf = model_selection.KFold(n_splits = k,shuffle = True)
```

```python
X_train, X_test, Y_train, Y_test, Y_pred = [0]*k, [0]*k, [0]*k, [0]*k, [0]*k
i = 0
for train, test in kf.split(Xt):
    X_train[i], X_test[i], Y_train[i], Y_test[i] = Xt[train], Xt[test], Y[train], Y[test]
    clf_nn.fit(X_train[i], Y_train[i])
    Y_pred[i] = clf_nn.predict(X_test[i])
    i+=1
```

```python
fig = plt.figure(figsize=[30,30])
i = 0
for i in range(k):
    plt.subplot(5,1,i+1)
    plt.scatter(Y_pred[i], Y_test[i])
    plt.xlabel('Prediction for y')
    plt.ylabel('Real y from test set')
    i+=1
```
